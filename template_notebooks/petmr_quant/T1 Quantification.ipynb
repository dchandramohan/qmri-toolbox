```python
# Compute T1 within each ROI

fa1, fa2 = np.deg2rad((8, 44))
TR = 11000

for roi in (rpIoh.roi_info):
    roi['fa'] = []
    roi['signal_magnitude'] = []
    roi['vox_T1'] = []
    
    for zi in range(nz):
        for yi in range(ny):
            for xi in range(nx):
                if (roi['mask'][xi, yi, zi]):
                    S1 = np.float(np.abs(np.complex(*recon_im_fa8[zi, yi, xi])))
                    S2 = np.float(np.abs(np.complex(*recon_im_fa44[zi, yi, xi])))
                    
                    T1 = -TR / np.log((S1/np.sin(fa1) - S2/np.sin(fa2)) / (S1/np.tan(fa1) - S2/np.tan(fa2)))
                    if not np.isnan(T1):
                        roi['fa'].extend((fa1, fa2))
                        roi['signal_magnitude'].extend((S1, S2))
                        roi['vox_T1'].append(T1)
    
    roi['fa'] = np.array(roi['fa'], dtype=np.float)
    roi['signal_magnitude'] = np.array(roi['signal_magnitude'], dtype=np.float)
    roi['vox_T1'] = np.array(roi['vox_T1'], dtype=np.float)
    
    print('ROI: {:s}\n\t{:d} datapoints...\n\tAvg. T1 = {:10.2f} msec\n'.format(roi['label'], roi['fa'].size, np.mean(roi['vox_T1'])/1000.0))
```

```python
# Water relaxivity vs. Gd concentration
water_idx = np.array([9, 17, 15, 6, 1, 3, 12], dtype=np.uint8)

Gd_water_t1s = np.array(list(np.mean(rpGd.roi_info[ci]['vox_T1'])/1000.0 for ci in water_idx))
Gd_water_conc = np.array([0.001, 0.002, 0.064, 0.032, 0.016, 0.008, 0.004])

fit = np.polyfit(Gd_water_conc, 1.0/Gd_water_t1s, 1)
fit_fn = np.poly1d(fit)

cc = np.linspace(0, .07, 10)
plt.close()

fig = plt.figure(figsize=(8.0,5.0))
plt.plot(Gd_water_conc, 1.0/Gd_water_t1s, 'rx')
plt.plot(cc, fit_fn(cc), 'b:')

plt.title('Water Relaxivity ({:s})'.format(dd))
plt.xlabel('Gd % by wt.')
plt.ylabel('R1 (ms)')

fig.set_facecolor('white')
fig.savefig('{:s}.Gd-WaterRlx.png'.format(dd))

plt.show()
```
